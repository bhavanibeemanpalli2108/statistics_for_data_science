```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
data=pd.read_csv("loan_approved.csv")
mean=data['ApplicantIncome'].mean()
std=data['ApplicantIncome'].std()
range=data['ApplicantIncome'].max()-data['ApplicantIncome'].min()
print("Mean:",mean)
print("Standard Deviation:",std)    
print("Range:",range)
```

```python
d1=data['ApplicantIncome'].head(5)
d1
```

```python
for idx, value in d1.items():
    z_score = (value - mean) / std
    print(f"Index: {idx}, Income: {value}, Z-Score: {z_score}")
```

# PREPROCESSING NUMERICAL DATA- TRANSFORMERS

```python
data['LoanAmount'].kurtosis()
data['LoanAmount'].skew()
```

```python
from sklearn.preprocessing import PowerTransformer
trans=PowerTransformer(method='yeo-johnson')
la_trans=trans.fit_transform(data[['LoanAmount']])
la_trans.shape
#boxcox method only for positive values
#yeo-johnson method for both positive and negative values
#here LoanAmount has only positive values so we can use both methods
```

```python
print("skewness:",pd.Series(la_trans.reshape(614,)).skew())
print("kurtosis:",pd.Series(la_trans.reshape(614,)).kurtosis())
# [[]] instead of [] use this for reshaping because fit_transform returns 2D array
# where skewness function works on 1D array , so we reshape it to 1D array
# that's why we use reshape function
```
